# python libraries used in this assignment

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import string
import re
```

# Data Extraction And Data Analysis

```python

# Read the input file
df = pd.read_excel("C:\\Users\\Naresh\\Downloads\\Input.xlsx")

# Create lists to store the computed values
positive_scores = []
negative_scores = []
polarity_scores = []
subjectivity_scores = []
avg_sentence_lengths = []
percentage_complex_words = []
fog_indices = []
avg_words_per_sentence = []
complex_word_counts = []
word_counts = []
syllables_per_word = []
personal_pronouns = []
avg_word_lengths = []

# Function to count the number of personal pronouns
def count_personal_pronouns(text):
    personal_pronoun_count = sum(1 for word in text.split() if word.lower() in ['i', 'we', 'my', 'ours', 'us'])
    return personal_pronoun_count
#This the list of stop words in nltk
custom_stopwords = [
    'a', 'about', 'above', 'after', 'again', 'against', 'ain', 'all', 'am', 'an', 'and', 'any', 'are',
    'aren', "aren't", 'as', 'at', 'be', 'because', 'been', 'before', 'being', 'below', 'between', 'both',
    'but', 'by', 'can', 'couldn', "couldn't", 'd', 'did', 'didn', "didn't", 'do', 'does', 'doesn', "doesn't",
    'doing', 'don', "don't", 'down', 'during', 'each', 'few', 'for', 'from', 'further', 'had', 'hadn',
    "hadn't", 'has', 'hasn', "hasn't", 'have', 'haven', "haven't", 'having', 'he', 'her', 'here', 'hers',
    'herself', 'him', 'himself', 'his', 'how', 'i', 'if', 'in', 'into', 'is', 'isn', "isn't", 'it', "it's",
    'its', 'itself', 'just', 'll', 'm', 'ma', 'me', 'mightn', "mightn't", 'more', 'most', 'mustn', "mustn't",
    'my', 'myself', 'needn', "needn't", 'no', 'nor', 'not', 'now', 'o', 'of', 'off', 'on', 'once', 'only',
    'or', 'other', 'our', 'ours', 'ourselves', 'out', 'over', 'own', 're', 's', 'same', 'shan', "shan't",
    'she', "she's", 'should', "should've", 'shouldn', "shouldn't", 'so', 'some', 'such', 't', 'than',
    'that', "that'll", 'the', 'their', 'theirs', 'them', 'themselves', 'then', 'there', 'these', 'they',
    'this', 'those', 'through', 'to', 'too', 'under', 'until', 'up', 've', 'very', 'was', 'wasn', "wasn't",
    'we', 'were', 'weren', "weren't", 'what', 'when', 'where', 'which', 'while', 'who', 'whom', 'why',
    'will', 'with', 'won', "won't", 'wouldn', "wouldn't", 'y', 'you', "you'd", "you'll", "you're", "you've",
    'your', 'yours', 'yourself', 'yourselves'
]
# Iterate over each row in the dataframe
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']

    # Send a GET request to fetch the webpage content
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the article title
    title = soup.find('title').text
    
    # Find the article text
    article_text = soup.get_text(strip=True)
    
    # Perform text analysis and compute the variables
    document_path ="C:\\Users\\Naresh\\Downloads\\positive-words.txt"  
    with open(document_path, "r") as file:
        document_content = file.read()
    document_words = set(document_content.split())
    count = 0
    for word in  article_text.split():
        if word in document_words:
            count+=1
    count
    positive_scores.append(count)
    
    # Negative Score
    document_path ="C:\\Users\\Naresh\\Downloads\\negative-words.txt"  
    with open(document_path, "r") as file:
        document_content = file.read()
    document_words = set(document_content.split())
    count1 = 0
    for word in  article_text.split():
        if word in document_words:
            count1+=1
    count1
    negative_scores.append(count1)
    
    # Polarity Score
    polarity_score = (count - count1) / (count + count1 + 0.000001)
    polarity_scores.append(polarity_score)
    
    # Subjectivity Score
    total_words = len(article_text.split())
    subjectivity_score = (count + count1) / (total_words + 0.000001)
    subjectivity_scores.append(subjectivity_score)
    
    # Average Sentence Length
    sentences = article_text.split('. ')
    avg_sentence_length = sum(len(sentence.split()) for sentence in sentences) / len(sentences)
    avg_sentence_lengths.append(avg_sentence_length)
    
    # Percentage of Complex Words
    complex_word_count = sum(1 for word in article_text.split() if word.strip(string.punctuation) and len(word) > 2)
    percentage_complex_words.append(complex_word_count / total_words * 100)
    
    # Fog Index
    fog_index = 0.4 * (avg_sentence_length + (complex_word_count / total_words * 100))
    fog_indices.append(fog_index)
    
    # Average Number of Words Per Sentence
    avg_words_per_sentence.append(total_words / len(sentences))

    
    # Word Count
    clean_text = re.sub(r'[^\w\s]', '', article_text)
    clean_text =  clean_text .split()
    list_without_stopwords = []
    for word in  clean_text :
        if word not in custom_stopwords:
            list_without_stopwords+=[word]
    word_counts.append(len(list_without_stopwords))
    
    # Syllable Count Per Word
    def count_syllables(word):
        word = re.sub(r'(es|ed)$', '', word, flags=re.IGNORECASE)
        syllable_count = len(re.findall(r'[aeiouy]+', word, flags=re.IGNORECASE))
        return syllable_count

    # Split the text into words
    words = re.findall(r'\b\w+\b', article_text)
    # Count the number of syllables in each word
    syllable_counts = [count_syllables(word) for word in words]
    complex_word_count=0
    # Print the word and its corresponding syllable count
    for word, syllables in zip(words, syllable_counts):
        if syllables>2:
            complex_word_count+=1

    syllables_per_word.append(syllable_counts)  
        
    # Complex Word Count
    complex_word_counts.append(complex_word_count)
    
    
    # Personal Pronouns
    personal_pronoun_count = count_personal_pronouns(article_text)
    personal_pronouns.append(personal_pronoun_count)
    
    # Average Word Length
    total_word_length = sum(len(word) for word in article_text.split() if word.strip(string.punctuation))
    avg_word_length = total_word_length / (total_words + 0.000001)
    avg_word_lengths.append(avg_word_length)

# Create a new dataframe to store the computed values
output_df = pd.DataFrame({
    'URL_ID': df['URL_ID'],
    'URL': df['URL'],
    'Title': title,
    'Positive Score': positive_scores,
    'Negative Score': negative_scores,
    'Polarity Score': polarity_scores,
    'Subjectivity Score': subjectivity_scores,
    'Average Sentence Length': avg_sentence_lengths,
    'Percentage of Complex Words': percentage_complex_words,
    'Fog Index': fog_indices,
    'Average Number of Words Per Sentence': avg_words_per_sentence,
    'Complex Word Count': complex_word_counts,
    'Word Count': word_counts,
    'Syllables Per Word': syllables_per_word,
    'Personal Pronouns': personal_pronouns,
    'Average Word Length': avg_word_lengths
})
```
